Boolean Logic

```python
weatherIsNice = True
haveUmbrella = False

if (not haveUmbrella) or (not weatherIsNice):
    print('Stay inside')
else:
    print('Go for a walk')    
```

```python
weatherIsNice = True
haveUmbrella = False

if not haveUmbrella or weatherIsNice:
    print('Go for a walk')
else:
    print('Stay inside')    
```
