**Control Flow**

**If / Else statments**

```python
a = True
if a:
    print('It is true!')
    print('Also print this')
else:
    print('It is false!')
print('Always print this')    
```

```python
a = True
b = False
c = True
if a:
    print('It is true!')
    print('Also print this')
    if b:
        print('Both are true')
        if c:
            print('All three are true')
else:
    print('It is false!')
print('Always print this')             
```
